```python
from IPython.display import display, clear_output
from urllib.request import urlopen
import pandas as pd
import datetime
import requests
import sched
import time
import json
from pandas_datareader import data
import fix_yahoo_finance as yf # yahoo專用的拿來拉股票資訊
import matplotlib.pyplot as plt # 繪圖專用
import talib #技術分析專用
pd.core.common.is_list_like = pd.api.types.is_list_like
s = sched.scheduler(time.time, time.sleep)

def stock_crawler(data):
    stock_list = ['00692']

    userstock_00692='00692'
    start = datetime.datetime.now() - datetime.timedelta(days=365) #先設定要爬的時間
    end = datetime.date.today()


    # 與yahoo請求
    pd.core.common.is_list_like = pd.api.types.is_list_like
    yf.pdr_override()

    # 取得股票資料
    stock_00692 = data.get_data_yahoo(userstock_00692+'.TW', start, end)

    # 取得00692的MA
    ret_00692 = pd.DataFrame(talib.SMA(stock_00692['Close'].values,5), columns= ['5-day average']) #10日移動平均線
    ret_00692 = pd.concat([ret_00692,pd.DataFrame(talib.SMA(stock_00692['Close'].values,10), columns= ['10-day average'])], axis=1) #10日移動平均線
    ret_00692 = pd.concat([ret_00692,pd.DataFrame(talib.SMA(stock_00692['Close'].values,20), columns= ['20-day average'])], axis=1) #20日移動平均線
    ret_00692 = pd.concat([ret_00692,pd.DataFrame(talib.SMA(stock_00692['Close'].values,60), columns= ['60-day average'])], axis=1) #60日移動平均線
    ret_00692 = ret_00692.set_index(stock_00692['Close'].index.values)

    five_day_avg = round(ret_00692['5-day average'].tail()[-1],2)
    ten_day_avg = round(ret_00692['10-day average'].tail()[-1],2)
    month_avg = round(ret_00692['20-day average'].tail()[-1],2)
    season_avg = round(ret_00692['60-day average'].tail()[-1],2)

    # 組成stock_list
    stock_list = '|'.join('tse_{}.tw'.format(target) for target in stock_list) 

    #　query data
    query_url = "http://mis.twse.com.tw/stock/api/getStockInfo.jsp?ex_ch="+ stock_list
    data = json.loads(urlopen(query_url).read())

    # 過濾出有用到的欄位
    columns = ['c','n','z','tv','v','o','h','l','y']
    df = pd.DataFrame(data['msgArray'], columns=columns)
    df.columns = ['股票代號','公司簡稱','當盤成交價','當盤成交量','累積成交量','開盤價','最高價','最低價','昨收價']
    df.insert(9, "漲跌百分比", 0.0) 

    # 新增漲跌百分比
    for x in range(len(df.index)):
        if df['當盤成交價'].iloc[x] != '-':
            df.iloc[x, [2,3,4,5,6,7,8]] = df.iloc[x, [2,3,4,5,6,7,8]].astype(float)
            df['漲跌百分比'].iloc[x] = (df['當盤成交價'].iloc[x] - df['昨收價'].iloc[x])/df['昨收價'].iloc[x] * 100

    # 紀錄更新時間
    time = datetime.datetime.now()  
    print("更新時間:" + str(time.hour)+":"+str(time.minute))
    print("股票代號:{}".format(userstock_00692))

    if df['當盤成交價'][0] < five_day_avg:
        print('跌破5日線')
        if df['當盤成交價'][0] < ten_day_avg:
            print('跌破10日線')
        if df['當盤成交價'][0] < month_avg:
            print('跌破月線')
        if df['當盤成交價'][0] < season_avg:
            print('跌破季線')
    else:
        print('在5、10、20、60日均線之上')

    start_time = datetime.datetime.strptime(str(time.date())+'9:30', '%Y-%m-%d%H:%M')
    end_time =  datetime.datetime.strptime(str(time.date())+'13:30', '%Y-%m-%d%H:%M')

    # 判斷爬蟲終止條件
    if time >= start_time and time <= end_time:
        s.enter(1, 0, stock_crawler, argument=(data,))


s.enter(5, 2, stock_crawler, argument=(data,))
s.run()
```
